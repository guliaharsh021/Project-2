```python
import mysql.connector

# To create connection with MySQL
connection = mysql.connector.connect(
    host='127.0.0.1',
    user='root',
    password='MySQL@4321'
)

# To create a DataBase
cursor = connection.cursor()
cursor.execute("CREATE DATABASE IF NOT EXISTS sample_database")

# To create a table in a DataBase
cursor.execute("USE sample_database")
cursor.execute("""
    CREATE TABLE IF NOT EXISTS INSTRUCTOR (
        ID INTEGER PRIMARY KEY NOT NULL,
        FNAME VARCHAR(20),
        LNAME VARCHAR(20),
        CITY VARCHAR(20),
        CCODE CHAR(2)
    );
""")

# Insert the data into the table
cursor.execute("""
    INSERT INTO INSTRUCTOR (ID, FNAME, LNAME, CITY, CCODE) VALUES 
    (1, 'Rav', 'Ahuja', 'TORONTO', 'CA'),
    (2, 'Raul', 'Chong', 'Markham', 'CA'),
    (3, 'Hima', 'Vasudevan', 'Chicago', 'US');
""")

# Commit the transaction
connection.commit()

# Verify the insertion
cursor.execute("SELECT * FROM INSTRUCTOR")
rows = cursor.fetchall()
for row in rows:
    print(row)

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
from sqlalchemy import create_engine, text
import pandas as pd

# Connect to the MySQL server using an existing database
engine = create_engine("mysql+pymysql://root:MySQL%404321@127.0.0.1:3306/mysql")

# Create a new database
with engine.connect() as connection:
    connection.execute(text("CREATE DATABASE IF NOT EXISTS socioeconomic"))

# Connect to the newly created database
engine_new_db = create_engine("mysql+pymysql://root:MySQL%404321@127.0.0.1:3306/socioeconomic")

# Load the data from the CSV URL
df = pd.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')

# Write the DataFrame to the new database table
df.to_sql("chicago_socioeconomic_data", engine_new_db, if_exists='replace', index=False, method="multi")
```

```python
import sqlite3
import pandas as pd

# Connect to SQLite database (create if not exists)
conn = sqlite3.connect('socioeconomic.db')

# Load data from CSV into DataFrame
df = pd.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')

# Write DataFrame to SQLite database
df.to_sql(name='chicago_socioeconomic_data', con=conn, if_exists='replace', index=False)

# Execute SQL query and fetch results
query = "SELECT * FROM chicago_socioeconomic_data LIMIT 5;"
results = pd.read_sql_query(query, conn)

print(results)

# Close connection
conn.close()
```

```python
from sqlalchemy import create_engine, text
import pandas as pd

# Connect to the MySQL server using an existing database
engine = create_engine("mysql+pymysql://root:MySQL%404321@127.0.0.1:3306/mysql")

# Create a new database if it doesn't exist
with engine.connect() as connection:
    connection.execute(text("CREATE DATABASE IF NOT EXISTS FinalDB"))

# Connect to the newly created database
engine_new_db = create_engine("mysql+pymysql://root:MySQL%404321@127.0.0.1:3306/FinalDB")

# Read CSV data into DataFrames
df_1 = pd.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
df_2 = pd.read_csv('https://data.cityofchicago.org/resource/9xs2-f89t.csv')
df_3 = pd.read_csv('https://data.cityofchicago.org/resource/ijzp-q8t2.csv')

# Write DataFrames to MySQL database using the engine for the new database
df_1.to_sql(name='CENSUS_DATA', con=engine_new_db, if_exists='replace', index=False, method="multi")
df_2.to_sql(name='CHICAGO_PUBLIC_SCHOOLS', con=engine_new_db, if_exists='replace', index=False, method="multi")
df_3.to_sql(name='CHICAGO_CRIME_DATA', con=engine_new_db, if_exists='replace', index=False, method="multi")
```

```python
# Find the total number of crimes recorded in the CRIME table.

%sql SELECT COUNT(*) AS The_Total_Number_Of_Crimes_Recroded FROM CHICAGO_CRIME_DATA;
```

```python
# List community area names and numbers with per capita income less than 11000.

%sql SELECT ca, community_area_name FROM CENSUS_DATA WHERE per_capita_income_ > "11000";
```

```python
# List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 

%sql SELECT case_number FROM CHICAGO_CRIME_DATA WHERE description LIKE '%minor%';
```

```python
# List all kidnapping crimes involving a child?

%sql SELECT * FROM CHICAGO_CRIME_DATA WHERE PRIMARY_TYPE = "KIDNAPPING" AND DESCRIPTION LIKE "%CHILD%"
```

```python
# List the kind of crimes that were recorded at schools. (No repetitions)

%sql SELECT DISTINCT primary_type FROM CHICAGO_CRIME_DATA WHERE location_description LIKE '%SCHOOL%;
```

```python
# List the type of schools along with the average safety score for each type.

%sql SELECT elementary_or_high_school AS school_type, AVG(safety_score) AS average_safety_score FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY elementary_or_high_school;
```

```python
# List 5 community areas with highest % of households below poverty line

%sql SELECT community_area_name, percent_households_below_poverty FROM CENSUS_DATA ORDER BY percent_households_below_poverty DESC LIMIT 5;
```

```python
# Which community area is most crime prone? Display the coumminty area number only.
%sql SELECT community_area FROM CHICAGO_CRIME_DATA GROUP BY community_area ORDER BY COUNT(*) DESC LIMIT 1;
```

```python
# Use a sub-query to find the name of the community area with highest hardship index

%sql SELECT community_area_name FROM CENSUS_DATA WHERE hardship_index = (SELECT MAX(hardship_index) FROM CENSUS_DATA); 
 
```

```python
# Use a sub-query to determine the Community Area Name with most number of crimes?

%sql SELECT community_area_name FROM CENSUS_DATA WHERE ca = (SELECT community_area FROM CHICAGO_CRIME_DATA GROUP BY community_area ORDER BY COUNT(*) DESC LIMIT 1);
```
